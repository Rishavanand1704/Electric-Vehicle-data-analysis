```python
import pandas as pd
df = pd.read_csv('ev.csv')
df.head(20)
```

```python
df.isna().sum()
```

```python
df[df.duplicated()]
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
df.drop(columns=['efficiency*', 'fastcharge*', 'germany_price_before_incentives', 'netherlands_price_before_incentives','tow_hitch'], inplace = True)
```

```python
df.dtypes
```

```python
df.rename(columns={
    'title': 'brand',
    'battery': 'battery_kw',
    '0_-_100': '0-100',
    'range*': 'range_km',
    'uk_price_after_incentives': 'showroom_price',
    'towing_capacity_in_kg': 'weight_capacity_kg',
    'price-range': 'cost_per_km'
}, inplace=True)
```

```python
df
```

```python
conversion_rate = 1.25

# Function to clean the price strings
def clean_price(price_str):
    if isinstance(price_str, str):  # Ensure the value is a string
        # Remove unwanted characters: asterisk, currency symbols, and commas
        cleaned_str = price_str.replace('*', '').replace('£', '').replace(',', '').strip()
        return float(cleaned_str)  # Convert cleaned string to float
    return 0.0  # Return a default value if not a string

# Clean the price column and convert to numeric
df['price_numeric'] = df['showroom_price'].apply(clean_price)

# Convert from pounds to dollars
df['price_dollars'] = df['price_numeric'] * conversion_rate

# Format the dollar price with a dollar symbol
df['price_dollars_str'] = df['price_dollars'].apply(lambda x: f'${x:.2f}')  # Format to 2 decimal places
```

```python
df
```

```python
df.drop(columns=['showroom_price', 'price_numeric', 'price_dollars'], inplace = True)
```

```python
df.rename(columns={'price_dollars_str': 'price_dollars'}, inplace=True)
```

```python
df.rename(columns={'0-100': '_0_to_100'}, inplace=True)
```

```python
conversion_rate = 1.25

df['cost_per_km_$'] = df['cost_per_km'] * conversion_rate

# Format the dollar price with a dollar symbol
df['cost_per_km_$'] = df['cost_per_km_$'].apply(lambda x: f'${x:.2f}')  # Format to 2 decimal places
```

```python
df.drop(columns=['cost_per_km'], inplace=True)
```
